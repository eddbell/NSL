### Accept-Reject Algorithm

**Accept-Reject** Monte Carlo is a method used in numerical simulations to generate random samples from a target probability distribution. Here's how it works in a nutshell:

You have a target probability distribution that you want to sample from.
You choose a simpler, known probability distribution (for example a uniform distribution) that can envelope the target distribution, meaning it's never smaller than the target.
You generate random samples from the proposal distribution and calculate a ratio: the value of the target distribution divided by the value of the proposal distribution at that sample point.

Then you generate a random number between 0 and 1 and if the ratio is greater than the random number, you accept the sample as a valid sample from the target distribution. If not, you reject it.

### Importance Sampling Algorithm

Define the target distribution that you want to estimate properties of. Let's call this distribution P(x).

Choose a simpler "importance distribution" Q(x) that covers the same sample space as P(x), but has a higher probability in regions where P(x) contributes significantly to the property you want to estimate.

Generate N random samples from the importance distribution Q(x). Let's call these samples ${x_1, x_2, ..., x_N}$.

For each sample $x_i$, calculate the weight $w_i$ as the ratio of the target distribution $P(x_i)$ to the importance distribution $Q(x_i)$:
   
   $$w_i = \frac{P(x_i)}{Q(x_i)}$$

The mean of the weights adjust the estimated expectation or integral.

### Exercise 02.1

The goal of this exercise it to apply 2 different techniques for sampling distributions.

Integrated via Monte Carlo in 2 methods:
- Accept-reject method : sampling a uniform distribution in $[0,1]$.

- Importance sampling: computing the linear and parabola *important* distributions sampling the inverse of the comulatives.


The pictures show 2 estimations of the integral:

$$I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = 1$$

The estimations of $I$</span> and their uncertainties are show in function of the number of throws  $M$</font> = 10000 and number of blocks $N$</font> = 100.

```python
import matplotlib
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 22})
import numpy as np

M=10000
N=100
plt.figure(figsize=(12,8))
sum_prog_MCimpsam_retta, err_prog_MCimpsam_retta = np.loadtxt("./data/MC_IS_lin.dat", usecols=(0,1), delimiter=',', unpack='true')
sum_prog_MCimpsam_prabola, err_prog_MCimpsam_parabola = np.loadtxt("./data/MC_IS_parabola.dat", usecols=(0,1), delimiter=',', unpack='true')
sum_prog_MCsimple, err_prog_MCsimple = np.loadtxt("./data/MC_uniform.dat", usecols=(0,1), delimiter=',', unpack='true')

x = np.arange(100)*M/N
teo = np.ones(N);

plt.errorbar(x,sum_prog_MCsimple-teo,yerr=err_prog_MCsimple,label="MC via uniform sampling")
plt.errorbar(x,sum_prog_MCimpsam_retta-teo,yerr=err_prog_MCimpsam_retta,label="MC via Important Sampling (RETTA)")
plt.errorbar(x,sum_prog_MCimpsam_prabola-teo,yerr=err_prog_MCimpsam_parabola,label="MC via Important Sampling (PARABOLA)")
plt.legend();
plt.xlabel('#throws')
plt.ylabel('average-theory')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,8))

plt.errorbar(x,sum_prog_MCimpsam_retta-teo,yerr=err_prog_MCimpsam_retta,label="MC via Important Sampling (RETTA)")
plt.errorbar(x,sum_prog_MCimpsam_prabola-teo,yerr=err_prog_MCimpsam_parabola,label="MC via Important Sampling (PARABOLA)")
plt.legend();
plt.xlabel('#throws')
plt.ylabel('average-theory')
plt.grid(True)
plt.show()
```

### Exercise 02.2
Simulations about 3D Random Walks (RW) on a **cubic lattice (1)** and in the **continuum (2)**.

The picture show the euclidean central distant trend:  $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$. 

**(1)** On a cubic lattice with lattice constant  $a = 1$ at each discrete time the walker makes a forward or backward step of length equal to *a* in one of the 3 principal directions of the lattice: *x* , *y* or *z*. 

Each step the *direction* is chosen selecting a random integer between 1 and 3, while the *toward* is chosen selecting the sign with a random integer btween -1 and 1.

**(2)** At each discrete time the walker makes a step of length equal to $a=1$ along a random direction obtained by sampling uniformly the solid angle:  $\theta \in [0,\pi]$ and $\phi \in [0,2\pi]$ and computing the cartesian components.

Once one have the cartesian component the average radius is given by $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}} = \sqrt{x^2+y^2+z^2}$ . 

Moreover we observe trends identical behavior and a diffusive like $f(N)=k\sqrt{N}$.


```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

M=10000
N=100

def f(x,k):  # defining the fitting function
    return k * np.sqrt(x)


discrete, discrete_err = np.loadtxt("./data/RW_discrete.dat", usecols=(0,1), delimiter=',', unpack='true')
continuum, continuum_err = np.loadtxt("./data/RW_continuum.dat", usecols=(0,1), delimiter=',', unpack='true')

x = np.arange(1,101)*M/N

fig, (ax1, ax2) = plt.subplots(1, 2,figsize=(16,8))
fig.suptitle('Random Walk')

#plot error bars
ax1.errorbar(x, discrete, yerr = discrete_err, ls = "None", color = "b")
ax2.errorbar(x,continuum,yerr=continuum_err, ls = "None", color = "b")
#plot scatter plot
ax1.scatter(x, discrete, s = 2* discrete, marker = "h", color = "b",label="Discrete RW")
ax2.scatter(x,continuum,s = 2* continuum, marker = "h", color = "b",label="Continuum RW")


p_opt, p_cov = curve_fit(f, x, discrete)
y_fit = f(x,p_opt)
ax1.plot(x,y_fit,label="FIT",color ="r") # plotting fitted function
print("<FIT INFO> DISCRETE RW: k = ", p_opt[0]," +- ", p_cov[0][0])

p_opt, p_cov = curve_fit(f, x, continuum)
y_fit = f(x,p_opt)
ax2.plot(x,y_fit,label="FIT",color="r") # plotting fitted function
print("<FIT INFO> DISCRETE RW: k = ", p_opt[0]," +- ", p_cov[0][0])
ax1.set_xlabel('#throws');ax2.set_xlabel('#throws')
ax1.set_ylabel('Central distance (discrete RW)');ax2.set_ylabel('Central distance (continuum RW)')

ax1.grid(); ax2.grid()
ax1.legend(); ax2.legend();

plt.show()
```
