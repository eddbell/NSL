### Exercise 02.1


The pictures show 2 estimation of:

$$I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = 1$$

Integrated via Monte Carlo in two ways: sampling a uniform distribution in $[0,1]$ and using importance sampling. The second scenario use a linear approximation to stimate the integral.

The estimations of $I$</span> and their uncertainties are show in function of the number of blocks, $N$</font>.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

M=10000
N=100
sum_prog_MCimpsam_retta, err_prog_MCimpsam_retta = np.loadtxt("../dati/dati2/dati2.1.2+.dat", usecols=(0,1), delimiter=',', unpack='true')
sum_prog_MCimpsam_prabola, err_prog_MCimpsam_parabola = np.loadtxt("../dati/dati2/dati2.1.2.dat", usecols=(0,1), delimiter=',', unpack='true')
sum_prog_MCsimple, err_prog_MCsimple = np.loadtxt("../dati/dati2/dati2.1.1.dat", usecols=(0,1), delimiter=',', unpack='true')

x = np.arange(100)*M/N
teo = np.ones(N);

plt.errorbar(x,sum_prog_MCsimple-teo,yerr=err_prog_MCsimple,label="MC via uniform sampling")
plt.errorbar(x,sum_prog_MCimpsam_retta-teo,yerr=err_prog_MCimpsam_retta,label="MC via Important Sampling (RETTA)")
plt.errorbar(x,sum_prog_MCimpsam_prabola-teo,yerr=err_prog_MCimpsam_parabola,label="MC via Important Sampling (PARABOLA)")
plt.legend();
plt.xlabel('#throws')
plt.ylabel('average-theory')
plt.grid(True)
plt.show()
```

Silumations about 3D Random Walks (RW) on a cubic lattice (1) and in the continuum (2).
The picture show the central distant trend: $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$.
Moreover we observe trends identical behavior and a diffusive like $f(N)=k\sqrt{N}$.


```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

M=100000
N=100

def f(x,k):  # defining the fitting function
    return k * np.sqrt(x)


discrete, discrete_err = np.loadtxt("../dati/dati2/dati2.2.1.dat", usecols=(0,1), delimiter=',', unpack='true')
continuum, continuum_err = np.loadtxt("../dati/dati2/dati2.2.2.dat", usecols=(0,1), delimiter=',', unpack='true')

x = np.arange(1,101)*M/N

fig, (ax1, ax2) = plt.subplots(1, 2,figsize=(10,5))
fig.suptitle('Random Walk')
ax1.errorbar(x,discrete,yerr=discrete_err,label="discrete scenario")
ax2.errorbar(x,continuum,yerr=continuum_err,label="continuum scenario")

p_opt, p_cov = curve_fit(f, x, discrete)
y_fit = f(x,p_opt)
ax1.plot(x,y_fit,label="FIT") # plotting fitted function


p_opt, p_cov = curve_fit(f, x, continuum)
y_fit = f(x,p_opt)
ax2.plot(x,y_fit,label="FIT") # plotting fitted function

ax1.set_xlabel('#throws');ax2.set_xlabel('#throws')
ax1.set_ylabel('Spread');ax2.set_ylabel('Spread')

ax1.grid(); ax2.grid()
ax1.legend(); ax2.legend();

plt.show()
```
